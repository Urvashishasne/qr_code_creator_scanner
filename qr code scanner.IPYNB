```python
import qrcode as qr
from PIL import Image
```

```python
link ="https://leetcode.com/studyplan/leetcode-75/"
```

```python
q=qr.QRCode(version=1,error_correction=qr.constants.ERROR_CORRECT_H,box_size=10,border=4)
```

```python
q.add_data(link)
q.make(fit=True)
im=q.make_image(fill_color="black",back_color="white")
im
```
